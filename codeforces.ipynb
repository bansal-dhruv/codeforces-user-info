```python
import requests
import json
import time
import hashlib
run=True

while(run==True):
    handle=input('Enter your user handle : ')
    url="https://codeforces.com/api/user.info?handles="
    url=(url+handle)
    response = requests.get(url,auth=('rw','rw'))
    me=json.loads(response.content.decode('utf-8'))

    if(me["status"]=="FAILED"):
        print(" User not found \n Re-Enter userhandle ")    
    else:
        time=int(time.time())
        time=str(time)
        me=(me['result'][0])
        run=False
        if(me.get('firstname', None)):
            print('Name : ',me['firstName'],me['lastName'])
        if(me.get('rating', None)):
            print('Rating : ',me['rating'])
        if(me.get('maxRating', None)):
            print('RatingsStatus : ',me['rank'],'     MaxRating : ',me['maxRating'],'     MaxRank : ',me['maxRank'] )
        #print(me['avatar'])
        urlc="https://codeforces.com/api/contest.list?gym=false"
        response = requests.get(urlc,auth=('rw','rw'))
        contest=json.loads(response.content.decode('utf-8'))
        contest=contest["result"][0:5]
        print()
        print('             **UPCOMING CONTEST**')
        for i in (contest):
            if( i['phase']=='BEFORE' ):
                print('-> Contest : ',i['name'])


        xxx=input('Enter your codeforces key : ')
        yyy=input('Enter your codeforces secret : ')
        hashc='723456/user.friends?apiKey='+xxx+'&time='+time+'#'+yyy
        hashc=bytes(hashc,'utf-8')
        hashc=hashlib.sha512(hashc).hexdigest()
        urlf='http://codeforces.com/api/user.friends?apiKey='+xxx+'&time='+time+'&apiSig=723456'+str(hashc)
        #print(urlf)
        response = requests.get(urlf)
        me=json.loads(response.content.decode('utf-8'))
        me=(me['result'])
        
        print()
        print('                  **Friends**')
        for i in me:
            print('                ->',i,end='')
            url="https://codeforces.com/api/user.info?handles="
            url=(url+i)
            #print(url)
            response = requests.get(url,auth=('rw','rw'))
            m=json.loads(response.content.decode('utf-8'))
            m=(m["result"][0])

            if(abs(int(m['lastOnlineTimeSeconds'])-int(time))<280):
                print("\033[0;37;40m  ~ user is online \033[0m")
            else:
                print()

            if(not m.get('rating', None)):
                print('                                     ~ Ratings :  NOT RATED')
                continue
            print('                                     ~ Ratings : ',m['rating'],end=" | ")
            print('RatingsStatus : ',m['rank'])
```
